## data pre-processing

```python
df.isnull().sum()
```

```python
# Define a function for mean imputation with try and except block
def mean_imputation(column):
    try:
        # Attempt to calculate the mean
        column_mean = column.astype(float).mean()
        # Fill missing values with the mean
        column.fillna(column_mean, inplace=True)
    except ValueError:
        # If conversion to float fails, do nothing
        pass

# Apply mean imputation to all columns except 'Category'
for column in df.columns:
    if column != 'Category':
        mean_imputation(df[column])
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.head()
```

```python
df = df.drop(columns=['Link', 'Currency', 'Voucher','Title'])
```

```python
df.info()
```

```python
df['Sold Count'] = df['Sold Count'].str.replace('K', '000' )
df['Sold Count'] = df['Sold Count'].str.replace(' Sold', '')
df['Original Price'] = df['Original Price'].str.replace('Rs. ', '')
df['Rating in Stars'] = df['Rating in Stars'].str.replace('/5', '')
df['Delivery'] = df['Delivery'].str.replace('Free Delivery', '1')
```

```python
df.head()
```

```python

# Iterate through each column and print unique values
for column in df.columns:
    print(f"Unique values in column '{column}':")
    print(df[column].unique())
    print()
```

```python

df['Rating in Stars'] = df['Rating in Stars'].replace({'5-Apr': 5.4, '5-May': 5.5, '5-Mar': 5.3, '5-Jan': 5.1})

df['Original Price'] = df['Original Price'].replace({',': ""})
df['Current Price'] = df['Current Price'].replace({',': ""})

```

```python
df.sample(10)
```

```python

# Define a function to attempt to convert a value to float
def try_convert_to_float(value):
    try:
        return float(value)
    except (ValueError, TypeError):
        # If conversion fails, return NaN
        return float('nan')

# Iterate through each column
for column in df.columns:
    # Skip 'object' type columns
    if df[column].dtype == 'object' and column != 'Category':
        try:
            # Attempt to convert the column to float
            df[column] = df[column].apply(try_convert_to_float)
        except Exception as e:
            print(f"Error converting column '{column}' to float: {e}")

# Now, you can check the data types of each column after conversion
print(df.dtypes)
```

### Feature Enginering

```python
# Apply one-hot encoding to the 'Category' column

df = pd.get_dummies(df, columns=['Category'])
```

```python
df.shape
```

```python
df.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error

# Define features (independent variables) and target (dependent variable)
features = df.drop(columns=['Sold Count'])  # Excluding 'Sold Count' and 'Category'
target = df['Sold Count']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Initialize and train the model
model = XGBRegressor()
model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
